```python
import pandas as pd

orders = pd.read_csv("../data/olist_orders_dataset.csv")

print(orders.shape)
orders.head()
```

```python
print(orders.shape)
```

```python
order_items = pd.read_csv("../data/olist_order_items_dataset.csv")

print(order_items.shape)
order_items.head()
```

```python
items_con_orders = order_items.merge(
    orders,
    on='order_id',
    how='inner'
)

print(items_con_orders.shape)
items_con_orders.head()
```

```python
products = pd.read_csv("../data/olist_products_dataset.csv")
translation = pd.read_csv("../data/product_category_name_translation.csv")

print(products.shape)
print(translation.shape)
```

```python
products_traducidos = products.merge(
    translation,
    on='product_category_name',
    how='left'
)

print(products_traducidos.shape)
products_traducidos.head()
```

```python
# Merge final: items + orders + products traducidos
fact = (
    order_items
    .merge(orders, on='order_id', how='inner')
    .merge(products_traducidos, on='product_id', how='left')
)

print(fact.shape)
fact.head()
```

```python
payments = pd.read_csv("../data/olist_order_payments_dataset.csv")
print(payments.shape)

# Pagos totales por order
payments_agg = (
    payments
    .groupby('order_id', as_index=False)['payment_value']
    .sum()
)

fact = fact.merge(payments_agg, on='order_id', how='left')
print(fact.shape)
```

```python
fact['order_delivered_customer_date'] = pd.to_datetime(fact['order_delivered_customer_date'])
fact['order_estimated_delivery_date'] = pd.to_datetime(fact['order_estimated_delivery_date'])

# Flags — equivalente a tus columnas booleanas en fact_sales
fact['is_delivered'] = fact['order_status'] == 'delivered'
fact['is_canceled'] = fact['order_status'].isin(['canceled', 'unavailable'])
fact['is_on_time'] = (
    fact['order_delivered_customer_date'] <= fact['order_estimated_delivery_date']
)

# KPIs
total_orders = fact['order_id'].nunique()
gmv = fact['price'].sum().round(2)
revenue = fact['payment_value'].sum().round(2)
total_freight = fact['freight_value'].sum().round(2)
canceled_orders = fact[fact['is_canceled']]['order_id'].nunique()
delivered_orders = fact[fact['is_delivered']]['order_id'].nunique()
on_time_orders = fact[fact['is_on_time'] == True]['order_id'].nunique()

print(f"GMV:                 {gmv}")
print(f"Revenue:             {revenue}")
print(f"Total orders:        {total_orders}")
print(f"Cancellation rate:   {(canceled_orders / total_orders * 100).round(2)}%")
print(f"On time delivery:    {(on_time_orders / delivered_orders * 100).round(2)}%")
print(f"AOV:                 {(revenue / total_orders).round(2)}")
print(f"Total freight:       {total_freight}")
```
